# WRSC 2018 challenge results

Maps made with [Folium](https://folium.readthedocs.io/en/latest/).

```python
import folium
from glob import glob
from IPython.display import HTML, display
import os.path as osp
import pandas
import yaml
```

```python
calshot_latlon = [50.821116, -1.310633]
```

## Fleet race

```python
markers = pandas.read_csv('../fleet_race.csv', index_col=0)
```

```python
def plot_track(csv_file):
    boat_trace = pandas.read_csv(csv_file)
    
    # Create the map
    map_osm = folium.Map(location=calshot_latlon, zoom_start=17)
    
    # Add the marker locations
    for name, latlon in markers.iterrows():
        folium.features.Circle(latlon, radius=5, popup=name).add_to(map_osm)

    # Add the boat track
    latlons = [(row.latitude, row.longitude) for row in boat_trace.itertuples()]
    map_osm.add_child(folium.features.PolyLine(latlons))
    return map_osm
```

```python
for path in sorted(glob('*.csv')):
    attempt_name = osp.splitext(osp.basename(path))[0]
    map_osm = plot_track(path)
    
    display(HTML('<h3>{}</h3>'.format(attempt_name)))
    display(map_osm)
```
